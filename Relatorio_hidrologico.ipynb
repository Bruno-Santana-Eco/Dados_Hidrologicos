```python
import pandas as pd
import os

# Caminho da pasta onde os arquivos estão localizados
folder_path = 'C:/Users/Samsung/Documents/GitHub-Datas/Dataset_Dados_Hidrológicos'

# Lista para armazenar os resultados
resultados_periodos = []

# Iterar sobre todos os arquivos .csv na pasta
for file in os.listdir(folder_path):
    if file.endswith('.csv'):
        file_path = os.path.join(folder_path, file)
        
        # Abrir o arquivo CSV
        df = pd.read_csv(file_path, sep=';', encoding='latin1')
        
        # Aplicar as transformações no DataFrame
        df1 = df.iloc[13:].reset_index(drop=True)
        df1.columns = df1.iloc[0]
        df1 = df1[1:]

        # Converta a coluna 'Data' para o formato de data
        df1['Data'] = pd.to_datetime(df1['Data'], format='%d/%m/%Y')

        # Ordenar as datas
        df1 = df1.sort_values(by='Data')

        # Criar uma série de datas com a diferença de 1 mês
        data_range = pd.date_range(start=df1['Data'].min(), end=df1['Data'].max(), freq='MS')

        # Verificar quais datas estão faltando
        datas_faltando = data_range.difference(df1['Data'])

        # Identificar períodos completos
        periodos_completos = []
        inicio_periodo = df1['Data'].iloc[0]

        for i in range(1, len(df1)):
            if df1['Data'].iloc[i] != df1['Data'].iloc[i-1] + pd.DateOffset(months=1):
                fim_periodo = df1['Data'].iloc[i-1]
                periodos_completos.append((inicio_periodo, fim_periodo))
                inicio_periodo = df1['Data'].iloc[i]

        # Adicionar o último período
        periodos_completos.append((inicio_periodo, df1['Data'].iloc[-1]))

        # Armazenar o resultado com o nome do arquivo
        resultados_periodos.append({
            'arquivo': file,
            'periodos_completos': periodos_completos,
            'datas_faltando': datas_faltando.strftime('%d/%m/%Y').tolist()
        })

# Exibir os resultados para cada arquivo
for resultado in resultados_periodos:
    print(f"Arquivo: {resultado['arquivo']}")
    print("Períodos contínuos e completos:")
    for inicio, fim in resultado['periodos_completos']:
        print(f"Início: {inicio.strftime('%d/%m/%Y')} - Fim: {fim.strftime('%d/%m/%Y')}")
    print("\nMeses faltando:")
    print(resultado['datas_faltando'])
    print("-" * 50)
```
